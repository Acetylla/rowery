```python
import json
import os
import pandas as pd
import datetime as dt
import folium

path = os.listdir(path="C:/Users/sagal/Desktop/Studia/1SEM/INTOBL/rowery")

files = []

for name in path:
    time = dt.datetime(int(name[0:4]), int(name[4:6]), int(name[6:8]), int(name[9:11]), int(name[11:13]), int(name[13:15]))
    files.append((time, name))

files = sorted(files)
```

```python
def bikes_at_station(files, station_index):
    bikes_list = []
    for file in files:
        with open('C:/Users/sagal/Desktop/Studia/1SEM/INTOBL/rowery/' + file[1]) as jsons:
            data = json.load(jsons)
            if data == []:
                continue
            bikes = data[0]['places'][station_index]["bike_numbers"]
            if bikes == None: 
                bikes = "" #brak rowerów na stacji 
            bikes_list.append((bikes, file[0]))
            
    return bikes_list

```

```python
stations=[]  #też tylko do sprawdzania danych
#rozpoczynamy badania aktywności na stacjach - wypożyczone rowery oraz oddane na stacje
activity = []
taken_from_stations = []
given_to_stations = []

for w in range(len(df)-2):
    bike_list = bikes_at_station(files, w) #indeks stacji
    stations.append((w , bike_list))
    
    #Zliczanie wypożyczonych rowerów na stacji 'w'
    count = 0
    station_count = 0
    for n in range(len(bike_list)-1):
        bikes_then = bike_list[n][0].split(",")
        bikes_now = bike_list[n+1][0].split(",")
        
        for c in bikes_then:
            if c not in bikes_now:
                count += 1
        station_count += count
        count = 0
    taken_from_stations.append([w, station_count])
    

 #print(stations[200]) 

    #Zliczanie oddanych rowerów na stacji 'w'
    counter = 0
    stat_count2 = 0
    for i in range(1, len(bike_list)):
       bike_numbers = bike_list[i][0].split(",")
        
       for b in bike_numbers:
           if b not in bike_list[i - 1][0]:
               counter += 1
       stat_count2 += counter
       counter = 0
    given_to_stations.append([w,stat_count2])
    #print(given_to_stations[:])

    activity.append([w, given_to_stations[w][1]+taken_from_stations[w][1]])


#for z in range(len(activity)):
  #  print(activity[z])

print(taken_from_stations[])
```
